```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('video_links.csv')

# Remove duplicate "https://www.youtube.com" from the links
df['Video URL'] = df['Video URL'].str.replace('https://www.youtube.comhttps://www.youtube.com', 'https://www.youtube.com', regex=False)

# Save the cleaned DataFrame back to the CSV
df.to_csv('cleaned_video_links.csv', index=False)

# Display the cleaned DataFrame
print(df)
```

```python
df.duplicated(subset=['Video URL']).sum()
```
